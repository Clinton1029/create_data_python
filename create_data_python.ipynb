## 3. Using A Pandas Series To Create  A Dataframe


#### 3.1 Poultry Dataset Using Pandas Series




```python
breed = pd.Series(["Kienyeji", "Broiler", "Layers", "Kuroiler", "Sasso"])
age_weeks = pd.Series([12, 6, 20, 14, 10])
weight_kg = pd.Series([1.6, 1.5, 1.9, 2.0, 1.7])
eggs_per_week = pd.Series([5, 0, 6, 4, 7])
farm_location = pd.Series(["Kiambu", "Nakuru", "Nyeri", "Eldoret", "Machakos"])

# Put all Series into a list
series_list = [breed, age_weeks, weight_kg, eggs_per_week, farm_location]

# Convert to DataFrame by stacking columns
poultry_df = pd.DataFrame(series_list).T

# Assign column names
poultry_df.columns = ["Breed", "Age_Weeks", "Weight_kg", "Eggs_Per_Week", "Farm_Location"]

# Display
poultry_df.head(6)
```
